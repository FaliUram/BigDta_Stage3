```python
df_inicio = spark.read.table("annex_1_aplication")


# Visualizacion de prueba 
display(df_inicio.limit(10))

```

```python
# Limpieza de datos 
from pyspark.sql.functions import col, trim

# Se eliminan espacios en blanco alrededor de los valores
df_limpio = df_inicio.withColumn("fl_limpia", trim(col("value")))

# Filtramos el dataFrame para eliminar registros  vacios o nulos 
df_filtro = df_limpio.filter((col("fl_limpia") != "") & (col("fl_limpia").isNotNull()))

# Seleccionamos la columna procesada y renombramos 
df_final = df_filtro.select(col("fl_limpia").alias("value"))

# Mostramos los datos limpios y transformados 
print("Vista previa de los datos transformados")
display(df_final.limit(10))
```
